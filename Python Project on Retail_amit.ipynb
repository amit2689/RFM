## TASK 2

```python
df_task1['recency_deciles']= pd.qcut(df_task1['Duration'],10)
```

```python
df_task1
```

```python
df_task2 = df_task1.groupby('recency_deciles')['Total_Revenue'].sum()
df_task2 = pd.DataFrame(df_task2)
```

```python
df_task2
```

```python
df_task2['cum_sum']= df_task2['Total_Revenue'].cumsum()
```

```python
df_task2
```

```python
df_task2["total_revenue_across_deciles"] = df_task2['Total_Revenue'].sum()
```

```python
df_task2.head()
```

```python
df_task2['per_tot_rev']= (df_task2['cum_sum']/df_task2['total_revenue_across_deciles'])
```
